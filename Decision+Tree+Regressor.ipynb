```python
# -*- coding: utf-8 -*-


# Decision Tree Regression

# Importing Libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Importing Data set
dataset = pd.read_csv('C:/Users/marut/Desktop/Decision_Tree_Regression/Decision_Tree_Regression/Position_Salaries.csv')
X = dataset.iloc[:, 1:2].values # Spliting input data from dataset
Y = dataset.iloc[:, 2].values





# Fitting the Decision tree regression to the dataset
from sklearn.tree import DecisionTreeRegressor
reg = DecisionTreeRegressor(random_state = 0)
reg.fit(X,Y)

# preddicting a new result
y_pred = reg.predict(6.5)


# Visualising the Decision Tree Regfression results
plt.scatter(X, Y, color = 'red')
plt.plot(X, reg.predict(X), color = 'blue')
plt.title('Truth or Bluff (Descision Tree Regressor)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()

```

```python
# The above visualization is not suitable for Decision Tree regressor as the model is simply joining the dots and it doesnt make-
# sense also the model is non coninuous.
# For non continuois model we have to increase the resolution of the visuals.
X_grid = np.arange(min(X), max(X), 0.01)
X_grid = X_grid.reshape((len(X_grid),1))
plt.scatter(X, Y, color = 'red')
plt.plot(X_grid, reg.predict(X_grid), color = 'blue')
plt.title('Truth or Bluff (Descision Tree Regressor)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()

# Now we can see the step's in the visualization that shows the model is non continuous and the predictions are realistic.
```
